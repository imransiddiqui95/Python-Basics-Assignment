Question 2 : Explain the difference between List and Tuple in Python.



Answer: In Python list and Tuples are both used to store multiple items in asingle variable but they have some important differences as follows.

* - List is a collection of items which is mutable (changeable)

  - Tuple is a collection of items which is immutable (unchangeable)

* - List uses square brackets []
  - Tuple uses  round brackets ()

*   - List items can be added, removed or changed
    - Tuple items cannot be changed after creation

*   - List is used when data needs to be changed
    - Tuple is used when data should not be change

Question 3 : What is a function in Python? Why are functions useful?

Anwer: A function in Python is a set of statements grouped together to perform a specific task. Functions help in organizing the program into smaller and manageable parts. In Python, a function is created using the keyword def, followed by the function name and parentheses.

Functions allow us to call the same block of code multiple times without rewriting it, which makes programming easier and more efficient.

  - Functions are useful because:


 - Reusability of code,
 once a function is written, it can be used many times in the program.

 - Better readability,
functions make the program easy to understand by dividing it into small parts.

 - Easy debugging and testing,
errors can be found and corrected easily because the code is divided into functions.

 - Reduces repetition,
functions avoid writing the same code again and again, which saves time and effort.
